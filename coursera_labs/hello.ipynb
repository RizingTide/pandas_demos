```python
# Print  the "total_profit," and “month_number” columns only.

df_profit_bymonth = df[["month_number", "total_profit"]]
df_profit_bymonth
```

```python
# Read the total profit of all months and show it using the Bar plot. 
    # The generated Bar plot must include the following properties: 
        # X label name = Month number.
        # Y label name = Total profit.
        # Title = Company profit per month.

# df_profit_bymonth.set_index("month_number", inplace=True)
# ax = df_profit_bymonth.plot(kind = "bar", figsize =  (10, 6))
# ax.set_xlabel("Month Number")
# ax.set_ylabel("Total Profit")
# ax.set_title("Company Profit Per Month")
# plt.xticks(rotation = 0)
# plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x="month_number", y="total_profit", data=df_profit_bymonth, color="steelblue")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.ylim(top=500000)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title("Company Profit Per Month")
plt.grid(linestyle="--")
plt.show()
```

```python
# Read the total profit of all of the months, and show the line plot with the following style properties:
    # The generated line plot must include the following Style properties: 
        # The line style should be dotted and the line color should be red.
        # Show the legend at the lower-right location.
        # X label name = Month number
        # Y label name = Total profit.
        # Title = Company profit per month.
        # Add a circle marker; the line marker color = red. 
        # The line width should be 3.

plt.figure(figsize=(10,6))
sns.lineplot(x="month_number", y="total_profit", linestyle="dotted", color="red", linewidth=3, marker="o", markersize=6, markerfacecolor="black", markeredgecolor="red",data=df_profit_bymonth)
plt.xlabel("Month Number")
plt.xticks(df["month_number"])
plt.ylabel("Total Profit")
plt.ylim(100000, 500000)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.title("Company Profit Per Month")
plt.legend(labels= ["Profit Data Last Year"], loc = "lower right")
plt.show()
```

```python
# Print all of the product sales data and show it using a multi-line plot.
# Display the number of units sold per month for each product using multi-line plots. (i.e., show a separate Plotline for each product).

plt.figure(figsize=(10,6))
sns.lineplot(x="month_number", y="facecream", linewidth="2", marker="o", data=df)
sns.lineplot(x="month_number", y="facewash", linewidth="2", marker="o", data=df)
sns.lineplot(x="month_number", y="toothpaste", linewidth="2", marker="o", data=df)
sns.lineplot(x="month_number", y="bathingsoap", linewidth="2", marker="o", data=df)
sns.lineplot(x="month_number", y="shampoo", linewidth="2", marker="o", data=df)
sns.lineplot(x="month_number", y="moisturizer", linewidth="2", marker="o", data=df)
plt.xticks(df["month_number"])
plt.legend(labels = ["Facecream Sales Data", "Facewash Sales Data", "Toothpaste Sales Data", "Bathingsoap Sales Data", "Shampoo Sales Data", "Moisturizer Sales Data"])
plt.xlabel("Month Number")
plt.ylabel("Number of Units Sold")
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title("Units Sold Per Month For Each Product")
plt.show()
```

```python
# Read “bathingsoap” sales data for each month and show it using a scatter plot.
    # Add a grid to the plot. The grid line style should be “–“.

plt.figure(figsize=(10,6))
sns.scatterplot(x="month_number", y="bathingsoap", data=df)
plt.grid(linestyle = "--")
plt.xlabel("Month Number")
plt.xticks(df["month_number"])
plt.ylabel("Number of Units Sold")
plt.title("Bathing Soap Sales Data Per Month")
plt.legend(labels = ["Bathing Soap Sales Data"], loc="upper left")
plt.show()
```
